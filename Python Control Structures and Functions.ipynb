1. 1.	List
Concept
A list is an ordered and mutable collection in Python that can store multiple values of different data types.

Example

numbers = [10, 20, 30]

numbers.append(40)

print(numbers)

Questions:

1.	Write a program to calculate the sum of all elements in a list.

2.	Write a program to remove duplicate elements from a list.

3.	Write a program to find the largest and smallest element in a list.

4.	Write a program to reverse a list without using built-in functions.

5.	Write a program to count even and odd numbers in a list.



```python
nums = [34, 67, 89, 12, 55]
count = 0

for i in nums:
    if i > 50:
        count += 1

print("Numbers greater than 50:", count)
```

```python
nums = [5, 6, 5, 7, 8, 6]
num2 = []
for i in nums:
    if i not in num2:
        num2.append(i)
print(num2)
```

```python
nums = [45, 12, 89, 3, 27]
largest = nums[0]
smallest = nums[0]
for i in nums:
    if i > largest:
        largest = i
    if i < smallest:
        smallest = i

print("Largest:", largest)
print("Smallest:", smallest)
```

```python
nums = [5, 9, 1, 7]
print(max(nums))
print(min(nums))
```

```python
nums = [1, 2, 3]
rev = []
for i in nums:
    rev = [i] + rev
print(rev)
```

```python
nums = [10, 11, 12, 13]
even = 0
odd = 0
for i in nums:
    if i % 2 == 0:
        even = even + 1
    else:
        odd = odd + 1
print("Even:", even)
print("Odd:", odd)
```


5.	Functions
Concept

A function is a reusable block of code that performs a specific task and can return a result.

Example

def multiply(a, b):

    return a * b

print(multiply(4, 5))

Questions

1.	Write a function to check whether a number is prime.

2.	Write a function to calculate factorial of a number.

3.	Write a function to
find the maximum of three numbers.

4.	Write a function that returns the sum of elements of a list.

5.	Write a function to check whether a string is a palindrome.


```python
def prime(n):
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

print(prime(7))
```

```python
def fact(n):
    f = 1
    for i in range(1, n+1):
        f *= i
    return f

print(fact(5))
```

```python
def maximum(a,b,c):
    return max(a,b,c)

print(maximum(2,9,5))
```

```python
def maximum(a, b, c):
    m = a
    if b > m:
        m = b
    if c > m:
        m = c
    return m

print(maximum(7, 3, 9))
```

```python
def sum_list(l):
    total = 0
    for i in l:
        total += i
    return total

print(sum_list([1,2,3]))
```

```python
def palindrome(s):
    return s == s[::-1]

print(palindrome("madam"))
```

6.	Exception Handling


Concept

Exception handling is used
to handle runtime errors using try, except, else, and finally blocks.

Example

try:

    x = int("abc")

except ValueError:

    print("Conversion error")

Questions

1.	Write a program to handle division by zero error.

2.	Write a program to handle invalid user input.

3.	Write a program using multiple except blocks.

4.	Write a program using try-except-else.

5.	Create a custom exception to validate age greater than 18.


```python
try:
    a = int(input("Enter number: "))
    print(100 / a)
except ZeroDivisionError:
    print("Cannot divide by zero")
```

```python
try:
    x = int(input())
    y = int(input())
    print(x / y)
except ValueError:
    print("Invalid number")
```

```python
try:
    x = int("10")
    print(x)
except:
    print("Error")
finally:
    print("Finished")
```

```python
try:
    a = int(input("Enter number: "))
    b = int(input("Enter number: "))
    print(a / b)

except ValueError:
    print("Invalid input")

except ZeroDivisionError:
    print("Cannot divide by zero")
```

```python
try:
    num = int(input("Enter a number: "))
except ValueError:
    print("Invalid input")
else:
    print("Square is:", num * num)
```

```python
class AgeError(Exception):
    pass

try:
    age = int(input("Enter age: "))

    if age < 18:
        raise AgeError("Must be 18+")

    print("Eligible")

except AgeError:
    print("Age is less than 18")

except:
    print("Invalid input")
```

from functools import
reduce

numbers = [1, 2, 3, 4]

result = reduce(lambda x,
y: x + y, numbers)
print(result)

Questions

3.	Use reduce to find maximum element in a list.

```python

nums = [2, 4, 6, 8]
result = reduce(lambda x, y: x + y, nums)
print(result)
```

```python
nums = [1, 3, 5]
result = reduce(lambda x, y: x * y, nums)
print(result)
```

11.	filter()


Concept

The filter function filters elements from an iterable based on a condition.

Example

numbers = [1, 2, 3, 4, 5]

result = list(filter(lambda x: x % 2 == 0, numbers))

print(result)

Questions

1.	Use filter to find even numbers from a list.
2.	Use filter to find odd numbers from a list.
3.	Use filter to find numbers greater than 50 from a list.
4.	Use filter to extract vowels from a string.
5.	Use filter to remove empty strings from a list.



```python
nums = [3, 7, 9, 10, 12]
result = list(filter(lambda x: x % 3 == 0, nums))
print(result)
```

```python
nums = [5, -2, 8, -1, 0]
result = list(filter(lambda x: x < 0, nums))
print(result)
```

```python
words = ["cat", "elephant", "dog", "tiger"]
result = list(filter(lambda x: len(x) > 4, words))
print(result)
```

```python
text = "python"
result = list(filter(lambda x: x not in "aeiou", text))
print(result)
```

```python
nums = [0, 5, 0, 10, 15, 0]
result = list(filter(lambda x: x != 0, nums))
print(result)
```
